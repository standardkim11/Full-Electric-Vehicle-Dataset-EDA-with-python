# **2. Data Transform**

```python
df = df.rename(columns = {'VIN (1-10)': 'VIN','Clean Alternative Fuel Vehicle (CAFV) Eligibility' : 'Clean Alternative Fuel Vehicle Eligibility'})
```

```python
df.dropna(axis=0,inplace=True)
```

**Data Check**

```python
df.head()
```

# **3. Data Visualization**

```python
# Data for visualization
df2= df.copy()
```

**Top 10 Automobile Manufacturers**

```python
top10_make = df2.groupby(['Make']).size().reset_index(name='count').sort_values(by='count', ascending=False)[0:10]
```

```python
top10_make
```

```python
px.bar(top10_make, x ='Make', y ='count',title = 'Top 10 Automobile Manufacturers',text_auto=True)
```

----------

**Electric Vehicle Type**

```python
EVT = df2.groupby(['Electric Vehicle Type']).size().reset_index(name='count').sort_values(by='count', ascending=False)
```

```python
EVT
```

```python
fig = px.pie(EVT, values= 'count',names= 'Electric Vehicle Type',title = 'Electric Vehicle Type')
fig.update_traces(textposition='outside',textinfo='label + percent + value')
```

---------

**Clean Alternative Fuel Vehicle Eligibility Rate**

```python
CAFV = df2.groupby(['Clean Alternative Fuel Vehicle Eligibility']).size().reset_index(name='count').sort_values(by='count', ascending=False)
```

```python
CAFV
```

```python
fig = px.pie(CAFV, values= 'count',names= 'Clean Alternative Fuel Vehicle Eligibility',title = 'Clean Alternative Fuel Vehicle Eligibility Rate')
fig.update_traces(textposition='outside',textinfo='label + percent + value')
```

**Average fuel efficiency Top 10**

----------

**Tesla, HYUNDAI annual enrollment count**

```python
change_year = df.groupby(['Model Year','Make']).size().reset_index(name ='count')
```

```python
Tesla = change_year.loc[change_year['Make'] == 'TESLA',:].reset_index()
HYUNDAI = change_year.loc[change_year['Make'] == 'HYUNDAI',:].reset_index()
```

```python
TH = pd.concat([Tesla,HYUNDAI])
```

```python
TH
```

```python
px.line(TH, x= 'Model Year',y = 'count',markers =True, title = 'TESLA, HYUNDAI annual enrollment count', color= 'Make')
```
